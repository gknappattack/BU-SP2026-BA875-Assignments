# Model 1 Regression - Price ($) and SE_indicator

```python
# Regression using Price and Special Event indicator as predictors
import statsmodels.api as sm

# Add constant to feature matrix which includes price and se_indicator
X_model1 = sm.add_constant(dfjeans[["price","se_indicator"]].astype("int"))

# Fit the OLS model
olsjeans1 = sm.OLS(dfjeans.sales, X_model1).fit()
```

```python
# Print summary
olsjeans1.summary()
```

## Analysis of Model 1 Regression

### Coefficients

Price returned with a coefficient of -3.31 while se_indicator had a coefficient of 75.15. We can interpret this as saying for each 1 unit increase of price ($1) we expect a decrease of about 3.31 units of sales, which lines up with common sense that as price increases, people's willingness to buy decreases (depending on the product).

On the other hand, the 75.15 coefficient on se_indicator shows that when there is a special event (se_indicator = 1) then we should expect sales to increase by 75, which shows a strong effect on the final sales number when relevant.

### P Values 
Both independent variables returned with a P-value of 0.000 (rounded to 3 digits), which indicates statistical significance. This would allow us to reject the null hypothesis that these two independent variables are not at least signficantly correleated with sales, since they are under the 0.05 threshold for 95% confidence.

### Adjusted R^2 score

For this regression, we have an adjusted R^2 score of .961, which can be interpreted as all of our predictors account for 96.1% of the variation observed in the dependent variable sales. This is quite a high score and gives at least a strong signal that this model includes features that are going to be useful for predicting sales effectively. Additionally, since it is the adjusted score, even with the penalty for including more terms (not that 2 terms is a large number by any means) it still scores highly, indicating a strong model for prediction.

# Model 2 Regression - Price as a discount and SE_indicator

```python
# Regression using Price and Special Event indicator as predictors
import statsmodels.api as sm

# Create dummy variable for price as discount
dfjeans['price_discount'] = dfjeans.price / 135.00

# Add constant to feature matrix which includes price and se_indicator
X_model2 = sm.add_constant(dfjeans[["price_discount","se_indicator"]].astype("int"))

# Fit the OLS model
olsjeans2 = sm.OLS(dfjeans.sales, X_model2).fit()
```

```python
olsjeans2.summary()
```

## Analysis of Model 2 Regression

### Coefficients

In model 2, price_discount had a coefficient of -98.1 while se_indicator had 68.75. Similar to the first regression, these tell us how much change we can expect in sales when each variable moves by 1. In this case, neither of these independent variables actually move by 1 but have a maximum value of 1 and either equal 0 (se_indicator) or can range values between 0 and 1 (price_discount). 

se_indicator is the same as model 1, aside from the change in the coefficient where it is lower than the previous model, while price_discount has the effect where as the discount increases (the % of the original price goes down or in other words, price_discount goes closer to 0) the negative effect of the coefficient is reduced. It is more or less a different way of exhibiting the same behavior as the first regression; as the price goes down, sales goes up (or in this case are reduced less since it is a negative coefficient).

### P Values 
All independent variables returned with a P-value of 0.000 (rounded to 3 digits), which indicates statistical significance. This would allow us to reject the null hypothesis that these two independent variables are not at least signficantly correleated with sales, since they are under the 0.05 threshold for 95% confidence.

This result is the same as model 1, so both models have independent variables that are all statistically significant for sales, indicating no clear advantage in that regard of one model over the other.

### Adjusted R^2 score

For regression 2, we have an adjusted R^2 score of .957, which can be interpreted as all of our predictors account for 95.7% of the variation observed in the dependent variable sales. This is quite a high score and gives at least a strong signal that this model includes features that are going to be useful for predicting sales effectively.

This is just slightly lower than the r^2 term in model 1 and while it is not much worse, it at least gives us a way to potentially compare and rank our models over the other. Since we are not doing any sort of train/test splitting and predicting in this case, if r^2 is the only metric we have to use to compare the two models, there would be a slight preference for model 1, since it has better predictive power for the variance in sales over model 2 according to the slightly higher r^2 score.
